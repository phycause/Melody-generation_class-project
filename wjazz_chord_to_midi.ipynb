```python
import mingus.core.chords as chords
import sqlite3
import numpy as np

conn = sqlite3.connect('wjazzd.db')
c = conn.cursor()
melody = c.execute('SELECT eventid, onset, duration, pitch FROM melody where melid = 1').fetchall()
chord = c.execute('SELECT * FROM sections where type = "PHRASE" and melid = 1').fetchall()
# chords = c.execute('SELECT chord_changes FROM solo_info where melid = 3').fetchall()
form = c.execute('SELECT * FROM sections where type = "FORM" and melid = 1').fetchall()

# print(chord)
# print(form)

def build_scale(root_note):
    note_chart_sharp = ["C", "C#", "D", "D#", "E", "F","F#", "G", "G#", "A", "A#", "B"]
    note_chart_b = ["C", "Db", "D", "Eb", "E","F", "Gb", "G", "Ab", "A", "Bb", "B"]
    if root_note in note_chart_sharp:
        root_note_idx = note_chart_sharp.index(root_note)
        scale_on_note = note_chart_sharp[root_note_idx:]
        rest = note_chart_sharp[:root_note_idx]
        for rest_note in rest:
            scale_on_note.append(rest_note)
    else:
        root_note_idx = note_chart_b.index(root_note)
        scale_on_note = note_chart_b[root_note_idx:]
        rest = note_chart_b[:root_note_idx]
        for rest_note in rest:
            scale_on_note.append(rest_note)
    return scale_on_note
        

beats = c.execute('SELECT * FROM beats').fetchall() # where melid = 10
for w_chord in beats:
    if w_chord[6] != "":
        print(w_chord[6])
        if "7alt" in w_chord[6]: #
            print chords.from_shorthand(w_chord[6].replace("7alt", "67"))
        
      
        elif "79b13b" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("79b13b", "7b9"))
            fix_chord[2] = build_scale(w_chord[6].replace("79b13b", ""))[8]
            print fix_chord                      
        
        elif "j7911#" in w_chord[6]: #9-5
            fix_chord = chords.from_shorthand(w_chord[6].replace("j7911#", "9"))
            fix_chord[2] = build_scale(w_chord[6].replace("j7911#", ""))[6]
            fix_chord[3] = build_scale(w_chord[6].replace("j7911#", ""))[11]
            print fix_chord
            
            
        elif "j79" in w_chord[6]: #9-5
            fix_chord = chords.from_shorthand(w_chord[6].replace("j79", "M9"))
            fix_chord.pop(2)
            print fix_chord
            
        
        elif "j7" in w_chord[6]:
            print chords.from_shorthand(w_chord[6].replace("j7", "M7")) 
            
        elif "7913" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("7913", "13"))
            fix_chord.pop(2) #5
            move_note = fix_chord.pop(1) #3
            fix_chord.insert(3, move_note)
            print fix_chord
         #######################   
        elif "+7911#" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("+7911#", "9"))
            fix_chord[2] = build_scale(w_chord[6].replace("+7911#", ""))[6]
            print fix_chord
            break
        elif "7911#" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("7911#", "9"))
            fix_chord[2] = build_scale(w_chord[6].replace("7911#", ""))[6]
            print fix_chord
            
        elif "-79b" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("-79b", "m7"))
            fix_chord[1] = build_scale(w_chord[6].replace("-79b", ""))[1]
            print fix_chord
            
        elif "79b" in w_chord[6]: #
            print chords.from_shorthand(w_chord[6].replace("79b", "7b9"))
            
        elif "79#" in w_chord[6]: #
            print chords.from_shorthand(w_chord[6].replace("79#", "7#9"))
            
        elif "o7" in w_chord[6]: #
            print chords.from_shorthand(w_chord[6].replace("o7", "dim7"))
            
        elif "o" in w_chord[6]: #
            print chords.from_shorthand(w_chord[6].replace("o", "dim"))
            
        elif "sus79" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("sus79", "sus47"))
            add_note = build_scale(w_chord[6].replace("sus79", ""))[2]
            fix_chord.insert(1, add_note)
            print fix_chord

        elif "sus7" in w_chord[6]: #
            print chords.from_shorthand(w_chord[6].replace("sus7", "sus47"))
            
        elif "+7" in w_chord[6]: #
            print chords.from_shorthand(w_chord[6].replace("+7", "7+"))
            
        elif "79" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("79", "9"))
            fix_chord.pop(2) #5
            move_note = fix_chord.pop(1) #3
            fix_chord.append(move_note)
            print fix_chord
        elif "-69" in w_chord[6]:
            fix_chord = chords.from_shorthand(w_chord[6].replace("-69", "m6"))
            add_note = build_scale(w_chord[6].replace("-69", ""))[2]
            fix_chord.insert(1, add_note)
            print fix_chord
            
        else:
            print chords.from_shorthand(w_chord[6])
            
        


```
